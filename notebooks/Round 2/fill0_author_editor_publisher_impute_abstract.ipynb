```python
import json
import pandas as pd

# Load training data
with open('train.json', 'r') as f:
    train_data = json.load(f)

# Load test data
with open('test.json', 'r') as f:
    test_data = json.load(f)

# Convert to pandas DataFrame for easier manipulation
train_df = pd.DataFrame(train_data)
test_df = pd.DataFrame(test_data)
print(train_df.head(10))
```

```python
#fill change_names_numeric to with  
#adding a feature for editor 
train_df['editor_count'] = train_df['editor'].apply(lambda x: len(x) if x is not None else 0)
#print(train_df['editor_count'])

#adding a feature for author 
train_df['author_count'] = train_df['author'].apply(lambda x: len(x) if x is not None else 0)

#adding a feature for puplisher 
train_df['publisher_count'] = train_df['publisher'].apply(lambda x: x.count(",") + 1 if x is not None else 0)
#print(train_df['publisher_count'])


change_names_numeric = ["author_count", "editor_count", "publisher_count"]
train_df[change_names_numeric]
```

```python
import numpy as np
from fancyimpute import IterativeImputer

# Adding a feature for abstract length, giving nan to None
train_df['abstract_length'] = train_df['abstract'].apply(lambda x: len(str(x)) if x is not None else np.nan)

# Separate the columns to impute from the rest of the DataFrame
impute_abstract = train_df[["abstract_length"]]

# Define the imputer
imputer = IterativeImputer(max_iter=10, random_state=0)

# Fit and transform the data
imputed_data = imputer.fit_transform(impute_abstract)

# Update the original DataFrame with the imputed values
train_df["abstract_length"] = imputed_data

train_df["abstract_length"]
```
